```python
def bill(res):
    vote_list = []
    bill = res['results']['votes']
    bill_rang = list(range(0, len(bill)))
    for i in bill_rang:
        vote = {}
        for key, value in bill[i].items():
            if key == 'democratic' or key == 'republican' or key == 'independent' or key == 'total':
                for key1, value1 in value.items():
                    vote.update({f"{key} {key1}": value1})
            else:
                vote.update({key: value})
        vote_list.append(dict(vote))
    return vote_list
```
